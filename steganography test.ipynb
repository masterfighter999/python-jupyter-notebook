```python
encoded_img = img.flatten()
```

```python

def decode(img):
    msg = ""
    idx = 0
    while msg[-5:] != '[END]':
        bits = [bin(i)[-1] for i  in img[idx:idx+8]]
        print(bits)
        bits = ''.join(bits)
        msg += chr(int(bits,2))
        idx+=8
        if idx > img.shape[0]:
            print("No hidden message")
            break
    return msg
```
